## Exploratory data analysis and featurization

Our raw dataset consists of .mid files downloaded from kunstderfuge.com (https://www.kunstderfuge.com). We have two composers represented in the dataset, Bach and Debussy, and their .mid files are located in separate folders ("music_database/Midi_files_2/bach" and "music_database/Midi_files_2/debussy").

For a given musical composition, there may be multiple .mid files, each containing a separate performance of that composition. For example, there are two .mid files for Debussy's Arabesque No. 1: "arabesque_1_(c)dery.mid" and "arabesque_1_(c)oguri.mid". The .mid files are labeled as "name of composition"_"name of performer".mid. For each composer, we want to count how many .mid files are in the dataset, as well as how many distinct compostions there are.

```python
def count(composer):
    files = glob.glob("data/Midi_files_3/" + composer + "/*.mid")

    file_list = []
    comp_list = []
    for i in files:
        filename= i[i.find("\\"):]
        filename1 = i[i.find("\\"):i.rfind("_")]
        file_list.append(filename)
        if filename1 not in comp_list:
            comp_list.append(filename1)

    #print(file_list)
    #print(comp_list)
    print(composer)
    print("number of files:", len(file_list))
    print("number of compositions:", len(comp_list))
    
count('debussy')
count('bach')
```

This shows that the Bach dataset contains around 27% more files than the Debussy dataset. This a mild dataset imbalance.

Next let's take a look at the kind of data the .mid files contain. In order to visualize the data, we extract from the .mid file a list of note events. This list is like a musical score, a set of instructions that tells the midi player how to play a piece of music. The list contains every note in the piece, its start, duration, pitch, velocity, and other information (such as instrument). In this project we are only interested in the start, duration and pitch. The libfmp library allows us to plot the score in a "pianoroll" representation, a graph of note events with time as the x-axis and pitch as the y-axis.

```python
composer = 'debussy'
files = glob.glob("data/Midi_files_3/" + composer + "/*.mid")
fn = files[100]
score = libfmp.c1.midi_to_list(fn)

libfmp.c1.visualize_piano_roll(score, figsize=(8, 3), velocity_alpha=True)
```

Viewed as a piano roll, the score is essentially a 2D image. Convolutional neural networks (CNNs) are used all the time to classify images. Can we use a CNN to identify a score's composer based on the pixelated 'image' of the score?

It would be useful to have a way to generate a 2D pixelated image directly from the list of note events. That way we could easily control parameters like the resolution. The code below generates a 2D feature representation from the list of note events extracted from the .mid files.

```python
def create_rep(filename,start,short_score):    

    end = 9.99 + start

    time_slices=np.arange(start,end,0.01)

    rep = np.zeros((80,1000))
    for i in range(20,109):
        intervals = []
        for j in short_score:
            if j[2]==i:
                time_min = j[0]
                time_max = time_min + j[1]
                intervals.append([time_min,time_max])
 
        count = -1
        for k in time_slices:
            count = count + 1
            for l in intervals:
                if l[0] <= k <= l[1]:
                    rep[int(i-30),int(count)] = 1
                
    plt.imshow(rep,aspect='auto', origin='lower', interpolation='none')
    plt.xlabel("Time (centiseconds)")
    plt.ylabel("Pitch")
    plt.title("2D Feature Representation")
    plt.show()
    libfmp.c1.visualize_piano_roll(short_score, figsize=(8, 3), velocity_alpha=True)

composer = 'debussy'
files = glob.glob("data/Midi_files_3/" + composer + "/*.mid")
fn = files[100]
score = libfmp.c1.midi_to_list(fn)
#print(score)

start = 0

##Select note events that occur within specified interval##
short_score = []
end = start + 9.99
for i in score:
    if i[0] < end:
        short_score.append(i)



create_rep(fn,start,short_score)
```

One parameter we can control is the time step used to create the 2D feature representation. If we stick to a 2D representation with dimensions (80,1000), using a larger time step means a longer duration segment can be captured. However it also means that the timing and duration of note events are described with less precision. For our initial trials, we decided to use a time step of 0.01 s.

```python
def create_rep(input_data,filename,start,score,ts):    

    short_score = []
    end = 999*ts + start
    for i in score:
        if start <= i[0] < end:
            short_score.append(i)

    time_slices=np.arange(start,end,ts)
    
    rep = np.zeros((80,1000))
    for i in range(20,109):
        intervals = []
        for j in short_score:
            if j[2]==i:
                time_min = j[0]
                time_max = time_min + j[1]
                intervals.append([time_min,time_max])
 
        count = -1
        for k in time_slices:
            count = count + 1
            for l in intervals:
                if l[0] <= k <= l[1]:
                    #print("true",count)
                    rep[int(i-30),int(count)] = 1
    input_data.append(rep)

    return input_data

composer = 'debussy'
files = glob.glob("data/Midi_files_3/" + composer + "/*.mid")
fn = files[0]
score = libfmp.c1.midi_to_list(fn)
start_list = []
for note in score:
    start_list.append(note[0])
        
start = 0# min(start_list)
time_steps = [0.01,0.05,0.1]
input_data = []
for i in time_steps:
    input_data = create_rep(input_data,fn,start,score,i)
    
fig, axes = plt.subplots(1, 3, figsize=(24, 6))  # Create 1 row and 2 columns of subplots

for i, ax in enumerate(axes):
    if i < len(input_data):
        rep = input_data[i]
        ts = time_steps[i]  # Time step in seconds
        cax = ax.imshow(rep, aspect='auto', origin='lower', interpolation='none', extent=[0, 1000 * ts, 0, 80])
        ax.set_xlabel("Time (seconds)", fontsize=14)
        ax.set_ylabel("Pitch", fontsize=14)
        ax.set_title(f"Time step: {ts} s",fontsize=14)
     

# Show the plot
plt.tight_layout()
plt.show()
```

Now we produce the 2D feature representations for all the examples in the training data set. In order to increase the size of the dataset, we extract multiple (10) segments from each .mid file, increasing the number of training examples by a factor of 10. The different segments are labeled by their start time. 

We store the training dataset (consisting of features extracted from the .mid files) in a dictionary. The dictionary keys are labels with the following format: {composer}\\{midi file name}\\{start time}, for example bach\\bwv873_2_wtcii\\0.0. Taking 10 segments from each Midi file, there are a total of 4070 examples in the training data set. 

```python
def create_startlist(filename, num_segments):

    fn = filename
    begin_list = []
    score = libfmp.c1.midi_to_list(fn)

    for note in score:
        begin_list.append(note[0])
        
    start = min(begin_list)
    last = max(begin_list)
    start_list = []
    for i in range(num_segments):
        segment_start = start + ((last-start)/(num_segments+1))*i
        start_list.append(segment_start)
  
    return start_list, score

def create_rep(filename,input_data,start,score):    
   
    short_score = []
    end = 9.99 + start
    for i in score:
        if start <= i[0] < end:
            short_score.append(i)

    time_slices=np.arange(start,end,0.01)

    rep = np.zeros((80,1000))
    for i in range(20,109):
        intervals = []
        for j in short_score:
            if j[2]==i:
                time_min = j[0]
                time_max = time_min + j[1]
                intervals.append([time_min,time_max])
 
        count = -1
        for k in time_slices:
            count = count + 1
            for l in intervals:
                if l[0] <= k <= l[1]:
                    #print("true",count)
                    rep[int(i-30),int(count)] = 1
                #print("true")

    input_data.append(rep)
    
    return input_data

##Specify folder where MIDI files are located##
file_list_bach = glob.glob("data/Midi_files_3/bach/*.mid")
file_list_debussy = glob.glob("data/Midi_files_3/debussy/*.mid")
file_list = file_list_bach + file_list_debussy

##Specify number of samples to be made from each file##
num_segments = 10

data = []
label_list = []

##Loop through the MIDI files to create the input features##
for i in file_list:
    #print(i)
    filename = i    
    last_backslash_index = filename.rfind('files_3/')
    name = filename[last_backslash_index + 8:]
    name = re.sub(r'\.mid', '', name)
    #print(name)
    
    ##Create list of start times for the segments##
    start_list, score = create_startlist(filename, num_segments) 
    
    ##Make 2D feature representation for each segment##
    for start in start_list:
        label_list.append(name + "\\" + "{:.1f}".format(start))
        data = create_rep(filename,data,start,score)

print(len(data))
print(len(label_list))

##Create dictionary for data##
total_dict = {label_list[i]: data[i] for i in range(len(label_list))}
```

## Model Training

Now that we've create the features, we are ready to train a model. Before we do that however, we should create a test data set. We could make our test dataset by selecting random examples from our training data set, but in order to create a more rigorous test set and avoid data leakage, we should make sure that examples in our test set are not taken from the same pieces/midi files as those in our train set. Otherwise the test accuracy could be indicating whether the model has learned to recognize patterns unique to individual compositions rather than to the composer. In order to do this, we first randomly select compositions from our list of unique compositions and then extract all the segments associated with those compositions from the training data set.

```python
def count(comp_list,composer):
    files = glob.glob("data/Midi_files_3/" + composer + "/*.mid")

    file_list = []
    
    for i in files:
        filename= i[i.find("\\"):]
        filename1 = i[i.find("\\"):i.rfind("_")]
        file_list.append(filename)
        if filename1 not in comp_list:
            comp_list.append(filename1)

    return comp_list

comp_list = []
comp_list = count(comp_list,"bach")
comp_list = count(comp_list,"debussy")

random.seed(12)
test_comps = random.sample(comp_list,5)
print(test_comps)

test_dict = {key: value for key, value in total_dict.items() if any(sub in key for sub in test_comps)}
train_dict = {key: value for key, value in total_dict.items() if all(sub not in key for sub in test_comps)}

n_test = len(test_dict)
n_train = len(train_dict)
n_total = len(total_dict)
print("number of examples in test set:",n_test)
print("number of examples in train set:", n_train)
print(n_total)
```

Now that we have made our training and test datasets, we prepare the data for training. We use one hot encoding for the labels.

```python
# Transform input data into numpy array

data = list(train_dict.values())
train_X = np.stack(data)
train_X = train_X.reshape(n_train, 80,1000, 1)
train_X = train_X.astype('float32')
print(train_X.shape)
```

```python
# Prepare training data labels for training using one hot encoding

train_Y = [0 if 'bach' in key else 1 for key in train_dict.keys()]
train_Y = np.array(train_Y)
train_Y_one_hot = to_categorical(train_Y)
print(train_Y_one_hot.shape)
```

We also further split the training data into train and validation sets. During training, the model will be evaluated on the validation data after each epoch. This will help with hyperparameter tuning and identifying overfitting. 

```python
# Train-validation data split

train_X,valid_X,train_label,valid_label = train_test_split(train_X, train_Y_one_hot, test_size=0.2, random_state=10)

print(train_X.shape,valid_X.shape,train_label.shape,valid_label.shape)
```

As an initial trial, I am using a model with three convolutional layers (with 3X3 filters), and three max-pooling layers, followed by two densely-connected NN layers. Cross entropy loss and the Adam optimizer are used.

```python
#Set training parameters
batch_size = 100
epochs = 12
num_classes = 2

#Define CNN model
model = Sequential()
model.add(Conv2D(16, kernel_size=(5, 5),activation='linear',input_shape=(80,1000,1),padding='same'))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPooling2D((2, 2),padding='same'))
model.add(Conv2D(32, (5, 5), activation='linear',padding='same'))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
model.add(Conv2D(64, (5, 5), activation='linear',padding='same'))
model.add(LeakyReLU(alpha=0.1))                  
model.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
model.add(Flatten())
model.add(Dense(64, activation='linear'))
model.add(LeakyReLU(alpha=0.1))                  
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=tensorflow.keras.losses.binary_crossentropy, optimizer=tensorflow.keras.optimizers.Adam(),metrics=['accuracy'])
#model.summary()

checkpoint_filepath = 'models/model1'

model_checkpoint_callback = ModelCheckpoint(
    filepath=checkpoint_filepath,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)

#Train CNN model
train = model.fit(train_X,train_label,batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(valid_X, valid_label),callbacks=[model_checkpoint_callback])
```

Let's look at some learning curves to see how the performance of the model changed with training.

```python
# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot accuracy
axs[0].plot(train.history['val_accuracy'])
axs[0].plot(train.history['accuracy'])
axs[0].set_title('Model Accuracy')
axs[0].set_ylabel('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].legend(['Validation Accuracy','Train Accuracy'], loc='upper left')

# Plot loss
axs[1].plot(train.history['val_loss'])
axs[1].plot(train.history['loss'])
axs[1].set_title('Model Loss')
axs[1].set_ylabel('Loss')
axs[1].set_xlabel('Epoch')
axs[1].legend(['Validation Loss','Train Loss'], loc='upper left')

plt.tight_layout()

plt.show()
```

The learning curves show that the model has learned the training data quite well, almost perfectly predicting the labels. It is less accurate on the validation data, a sign of overfitting. Interestingly, after around 4-5 epochs, the validation loss is increasing with more training, while the model accuracy stays about the same. This shows that cross entropy and accuracy are not necessarily correlated.

In order to address the problem of overfitting, regularization can be used. We tried a combination of dropout and L2 regularization.

```python
from keras.regularizers import l2
#Set training parameters
batch_size = 100
epochs = 12
num_classes = 2

#Define CNN model
model_reg = Sequential()
model_reg.add(Conv2D(16, kernel_size=(5, 5),activation='linear',input_shape=(80,1000,1),padding='same',kernel_regularizer=l2(0.01)))
model_reg.add(LeakyReLU(alpha=0.1))
model_reg.add(MaxPooling2D((2, 2),padding='same'))
model_reg.add(Dropout(0.25))
model_reg.add(Conv2D(32, (5, 5), activation='linear',padding='same',kernel_regularizer=l2(0.01)))
model_reg.add(LeakyReLU(alpha=0.1))
model_reg.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
model_reg.add(Dropout(0.25))
model_reg.add(Conv2D(64, (5, 5), activation='linear',padding='same',kernel_regularizer=l2(0.01)))
model_reg.add(LeakyReLU(alpha=0.1))                  
model_reg.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
model_reg.add(Dropout(0.25))
model_reg.add(Flatten())
model_reg.add(Dense(64, activation='linear'))
model_reg.add(LeakyReLU(alpha=0.1))
model_reg.add(Dropout(0.5))
model_reg.add(Dense(num_classes, activation='softmax'))

model_reg.compile(loss=tensorflow.keras.losses.binary_crossentropy, optimizer=tensorflow.keras.optimizers.Adam(),metrics=['accuracy'])
#model.summary()

checkpoint_filepath = 'models/model1_reg'

model_checkpoint_callback = ModelCheckpoint(
    filepath=checkpoint_filepath,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)

#Train CNN model
train_reg = model_reg.fit(train_X,train_label,batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(valid_X, valid_label),callbacks=[model_checkpoint_callback])
```

```python
# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot accuracy
axs[0].plot(train_reg.history['val_accuracy'])
axs[0].plot(train_reg.history['accuracy'])
axs[0].set_title('Model Accuracy')
axs[0].set_ylabel('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].legend(['Validation Accuracy','Train Accuracy'], loc='upper left')

# Plot loss
axs[1].plot(train_reg.history['val_loss'])
axs[1].plot(train_reg.history['loss'])
axs[1].set_title('Model Loss')
axs[1].set_ylabel('Loss')
axs[1].set_xlabel('Epoch')
axs[1].legend(['Validation Loss','Train Loss'], loc='upper left')

plt.tight_layout()

plt.show()
```

We can also look more closely at the models performance by computing a confusion matrix. 


```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

ynew = reconstructed_model.predict(test_X)
classes_x=np.argmax(ynew,axis=1)
cm = confusion_matrix(test_Y, classes_x)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=np.array(('bach','debussy')))
disp.plot()
```

## Summary



We have trained a convolutional neural network model on 2D representations of musical excerpts taken from .mid files. The models show great promise in classifying the musical excerpts by composer.

Future work on this project includes further hyperparameter tuning, experimenting with different feature representations and model architectures, and exploring model interpretability. The 2D "piano-roll" feature representation was shown to be effective in this work. However, it has some drawbacks. It requires a large dimensionality in order to capture note events with the proper resolution. In addition, the feature representations are rather sparse; most elements are zero in any given example. Graph neural networks may be better suited to the continuous nature of the musical note events. 

One possible application of this work would to use it as a musicological aide for identifying unique characteristics of composer's styles. What specific patterns is the model memorizing in order to identify the composer. The model could be used to identify works of doubtful authenticity.
